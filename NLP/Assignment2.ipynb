## 1. Transformer's Attention Layer

We will first start with going over a few concepts that you learned in your high school statistics class. The variance of a random variable $X$, $\text{Var}(X)$ is defined as $\text{E}[(X-\mu)^2]$ where $\mu$ is the mean of $X$. Furthermore, given two independent random variables $X$ and $Y$ and a constant $a$,
$$ \text{Var}(X+Y) = \text{Var}(X) + \text{Var}(Y),$$
$$ \text{Var}(aX) = a^2\text{Var}(X),$$
$$ \text{Var}(XY) = \text{E}(X^2)\text{E}(Y^2) - [\text{E}(X)]^2[\text{E}(Y)]^2.$$

$$\text{Var}\left(\sum_i^n X_i Y_i\right) = n.$$

As the expectation of random variables are zero as stated in problem 1.1, I can express equation from above as below:

$$\frac{1}{d_k}  \begin{bmatrix} Var(x_{11})Var(y_{11}) + Var(x_{12})Var(y_{12}) + Var(x_{1, d_k})Var(y_{1, d_k})& Var(x_{11})Var(y_{21}) + Var(x_{12})Var(y_{22}) + Var(x_{1, d_k})Var(y_{2, d_k}) & \cdots  &  Var(x_{11})Var(y_{n1}) + Var(x_{12})Var(y_{n2}) + Var(x_{1, d_k})Var(y_{n, d_k})\\

When the variances for $Q$ and $K$ are higher than 1, it means that all of the elements of matrix above become bigger than one. It means that the gradient through the softmax function becomes extremely small, so it impedes the learing process. On the other hand, if the variances are less than 1, then all the values of matrix become less than one, and the problem which arises from variance being less is less problematic than being bigger than one.

````python
def preprocess(example):

  output = dict()
  context_part = example['context'].split(' ')
  question_part = example['question'].split(' ')

  output['context'] = context_part
  output['question'] = question_part

  if 'answers' not in example.keys():
    return output

  answer_list = []
  answer_candidate = example['answers']['text']

  for i, answer in enumerate(answer_candidate): # Several text elements exist in validation set. 

    temp_dict = dict()
    char_start_idx = example['answers']['answer_start'][i]
    csi_slicing = example['context'][:char_start_idx-1] # Because of the blank existing in front of slicing site
    answer_slicing = answer.split(' ')
    start_idx = len(csi_slicing.split(' '))
    end_idx = start_idx + len(answer_slicing) - 1

    temp_dict['start'] = start_idx
    temp_dict['end'] = end_idx
    answer_list.append(temp_dict)

  output['answers'] = answer_list

  return output

print('Just for Sanity Check: {}'.format(preprocess(squad_dataset['train'][0])['answers'])) # Done! Compare with squad_dataset['train'][0]['context'].split(' ')[90]
print('Just for Sanity Check: {}'.format(preprocess(squad_dataset['validation'][2])['answers'])) # Done! Compare with squad_dataset['validation'][0]['context'].split(' ')[28]
````

````python
def accuracy_estimate(dataset):  
  accuracy = []
  print('Print just for comparison!')
  print('---------------------------')

  for i, data in enumerate(dataset):
    original_text = data['answers']['text'][0]

    preprocessed = preprocess(data)
    start_idx = preprocessed['answers'][0]['start']
    end_idx = preprocessed['answers'][0]['end']
    concat_text = preprocessed['context'][start_idx:(end_idx+1)]
    concat_text = ' '.join(concat_text)

    if i <= 10:
      print('Original answer text:       {}'.format(original_text))
      print('Concatenated answer tokens: {}'.format(concat_text))
      print('--------------------------------')

    accuracy.append(original_text == concat_text)

  return sum(accuracy) / len(dataset)

print('Percentage of exact match: {}%'.format(accuracy_estimate(squad_dataset['train'])*100)) # Most of mismatch happen at puncuation level. 
````

````python
def get_token_bag(data, tokenizer_to_use, cq=False): # data: for example, squad_dataset['train']
  train_token_bowl = []

  for i in range(len(data)):
    
    if cq == True:
      token_q_instance = tokenizer_to_use.tokenize(data[i])
      train_token_bowl.extend(token_q_instance)

    else:
      token_c_instance = tokenizer_to_use.tokenize(data[i]['context'])  # token_c_instance: context instance
      train_token_bowl.extend(token_c_instance)

    # I put 'PaD', 'UnK', 'EoS' instead of '<pad>', '<unk>', '<EOS>' because of the tokenizer nominated by problem. 
    # Though I made a embedding as '<pad>', '<unk>', '<EOS>', when I use tokenier at the real vocabs, it is also tokenized.
    # For example, <pad> token inserted might be tokenized as '<' + 'pad' + '>', so they might not refer <pad> token in embedding. 
   
    if i == len(data)-1:
      train_token_bowl = list(set(train_token_bowl))
      train_token_bowl.insert(0, 'PaD')
      train_token_bowl.insert(1, 'UnK')
      if cq == True:
        train_token_bowl.remove('EoS') 
      train_token_bowl.insert(2, 'EoS')

  return train_token_bowl

def bag_word_to_index(token_list):
  return {word: index for index, word in enumerate(token_list)}

train_token_list = get_token_bag(squad_dataset['train'], tokenizer) # The list of tokens that once appeared on training set
token_c_idx = bag_word_to_index(train_token_list) # Mapping the tokens to index
````

````python
# Feature data in trainset, processed for TensorData

def train_feature_data_processing(data, which_tokenizer, which_token_idx):
  return_list = []
  data_batch = list(map(lambda x: which_tokenizer.tokenize(x), data))

  for instance in data_batch:
    instance_mapping = list(map(lambda x: which_token_idx[x], instance))
    return_list.append(torch.LongTensor(instance_mapping))

  return return_list

train_data_processed = train_feature_data_processing(squad_dataset['train']['context'], tokenizer, token_c_idx)
train_x_padded = pad_sequence(train_data_processed, batch_first=True, padding_value=0)
````

````python
# Feature data in validation set, processed for TensorData

def valid_unk_processing(data, which_tokenizer, whole_token, which_index): # To change 'unseen' words in the validation 
 
  return_list = []
  whole_token_array = np.array(whole_token)

  for _ , instance in enumerate(data):
    instance_array = np.array(which_tokenizer.tokenize(instance))
    idx = np.where(np.logical_not(np.isin(instance_array, whole_token_array)))
    instance_array[idx] = 'UnK'
    instance_array = torch.LongTensor(list(map(lambda x: which_index[x], instance_array.tolist())))
    return_list.append(instance_array)

  return return_list
````

````python
class LSTMModel(nn.Module):
  def __init__(self, vocab_size, embed_dim, hidden_dim=128, output_dim=1, bidirectional=False, dropout=0.5):
    super(LSTMModel, self).__init__()
    self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
    self.lstm = nn.LSTM(input_size=embed_dim, 
                        hidden_size=hidden_dim, 
                        bidirectional=bidirectional, 
                        dropout=dropout, 
                        batch_first=True)
    self.relu = nn.ReLU()

    if bidirectional == True:
      self.start_proj = nn.Linear(hidden_dim*2, output_dim) 
      self.end_proj = nn.Linear(hidden_dim*2, output_dim)
    else:
      self.start_proj = nn.Linear(hidden_dim, output_dim)
      self.end_proj = nn.Linear(hidden_dim, output_dim)

    self.dropout = nn.Dropout(dropout)

    nn.init.xavier_uniform_(self.start_proj.weight)
    nn.init.xavier_uniform_(self.end_proj.weight)

  def forward(self, embed_index):

    embed_lengths = torch.LongTensor([torch.max(embed_index[i,:].data.nonzero())+1 for i in range(embed_index.size(0))])
    embed_lengths, sorted_idx = embed_lengths.sort(0, descending=True)
    embed_index = embed_index[sorted_idx]
    embedded = self.embedding(embed_index)

    packed_input = pack_padded_sequence(embedded, embed_lengths, batch_first=True) 
    packed_output, _ = self.lstm(packed_input)
    output, _ = pad_packed_sequence(packed_output, batch_first=True)

    output = self.dropout(self.relu(output))
    start_output = self.start_proj(output)
    end_output = self.end_proj(output)

    return start_output, end_output
````

## 4. LSTM + Attention for SQuAD

**Problem 4.1** *(20 points)* Here, we will be appending an attention layer on top of LSTM outputs. We will use a single-head attention sublayer from Transformer. That is, you will implement 
$$ \text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^\top}{\sqrt{d}}\right)V,$$
where $Q, K, V$ is obtained by the linear transformation of the hidden states of the LSTM outputs $H$, i.e. $Q = HW^Q, K=HW^K, V=HW^V$ ($W^Q, W^K, W^V \in \mathbb{R}^{d \times d}$ are trainable weights). Note that the output of $\text{Attention}$ layer has the same dimension as $H$, so you can directly append your token classification layer on top of it. Report the accuracy and compare it with 3.2.

**Answer to Problem 4.1** I could feel the power of attention. By just adding the attention layer, accuracy jumped about 4 to 5 times comparing to previous ones. On my experiment, it recorded about 18% which was tremendous improvement than the previous ones. I did permit the output index to be emitted from the question part, thinking that machine would identify the role of `<eos>` token so that it could distinguish the different role of questions automatically  .



````python
class LSTMModelwithAttention(nn.Module):
  def __init__(self, vocab_size, embed_dim, hidden_dim=128, output_dim=1, bidirectional=False, dropout=0):
    super(LSTMModelwithAttention, self).__init__()

    self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
    self.lstm = nn.LSTM(input_size=embed_dim, hidden_size=hidden_dim, bidirectional=bidirectional, 
                        dropout=dropout, batch_first=True)
    self.hidden_dim = hidden_dim
    self.relu = nn.ReLU()
    
    hidden_cnt = 2 if bidirectional == True else 1
    self.start_proj = nn.Linear(hidden_dim * hidden_cnt, output_dim) 
    self.end_proj = nn.Linear(hidden_dim * hidden_cnt, output_dim)
    self.w_queue = nn.Linear(hidden_dim * hidden_cnt, hidden_dim * hidden_cnt)
    self.w_key = nn.Linear(hidden_dim * hidden_cnt, hidden_dim * hidden_cnt)
    self.w_value = nn.Linear(hidden_dim * hidden_cnt, hidden_dim * hidden_cnt)

    self.dropout = nn.Dropout(dropout)

    nn.init.xavier_normal_(self.start_proj.weight)
    nn.init.xavier_normal_(self.end_proj.weight)
    nn.init.xavier_normal_(self.w_queue.weight)
    nn.init.xavier_normal_(self.w_key.weight)
    nn.init.xavier_normal_(self.w_value.weight)

  def forward(self, embed_index):
    embed_lengths = torch.LongTensor([torch.max(embed_index[i,:].data.nonzero())+1 for i in range(embed_index.size(0))])
    embed_lengths, sorted_idx = embed_lengths.sort(0, descending=True)
    embed_index = embed_index[sorted_idx]
    embedded = self.embedding(embed_index)

    packed_input = pack_padded_sequence(embedded, embed_lengths, batch_first=True) 
    packed_output, _ = self.lstm(packed_input)
    hidden, _ = pad_packed_sequence(packed_output, batch_first=True)
    # hidden = self.relu(hidden)

    Q = self.w_queue(hidden).to(device)
    K = self.w_key(hidden).to(device)
    V = self.w_value(hidden).to(device)
    dim = self.hidden_dim
    
    scores = torch.matmul(Q, K.transpose(-2,-1)) / math.sqrt(dim) * 100
    # mask = np.triu(np.ones((1, scores.shape[1], scores.shape[2])), k=1).astype('uint8')
    # mask = torch.from_numpy(mask) == 0
    # scores = scores.to(device).masked_fill(mask.to(device) == 0, -1e9)
    attn = torch.matmul(self.dropout(F.softmax(scores, dim=-1)), V)
    start_output = self.start_proj(attn)
    end_output = self.end_proj(attn)

    return start_output, end_output
````

**Problem 4.2** *(10 points)* On top of the attention layer, let's add another layer of (bi-directional) LSTM. So this will look like a *sandwich* where the LSTM is bread and the attention is ham. How does it affect the accuracy? Explain why do you think this happens. 

**Answer to Problem 4.2** I got an accuracy about 1% which was lower than the LSTM-only version. There are two possibilities. First is because I permitted to predict indices including question part. Second is, though LSTM + Attention did well, by accumulating one more upper layer using LSTM, gradient vanished a lot.

````python
class LSTMSandwich(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim=128, 
                 output_dim=1, bidirectional=False, dropout=0, context_question=True):
        super(LSTMSandwich, self).__init__()

        hidden_cnt = 2 if bidirectional == True else 1
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(input_size=embed_dim, 
                            hidden_size=hidden_dim, 
                            bidirectional=bidirectional, 
                            dropout=dropout, 
                            batch_first=True)
        self.lstm2 = nn.LSTM(input_size=hidden_dim * hidden_cnt, 
                             hidden_size=hidden_dim, 
                             bidirectional=bidirectional, 
                             dropout=dropout, 
                             batch_first=True)
        self.hidden_dim = hidden_dim
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout)
        self.cq = context_question

        self.start_proj = nn.Linear(hidden_dim * hidden_cnt, output_dim) 
        self.end_proj = nn.Linear(hidden_dim * hidden_cnt, output_dim)
        self.w_queue = nn.Linear(hidden_dim * hidden_cnt, hidden_dim * hidden_cnt)
        self.w_key = nn.Linear(hidden_dim * hidden_cnt, hidden_dim * hidden_cnt)
        self.w_value = nn.Linear(hidden_dim * hidden_cnt, hidden_dim * hidden_cnt)

    def forward(self, embed_index):
        embed_lengths = torch.LongTensor([torch.max(embed_index[i,:].data.nonzero())+1 for i in range(embed_index.size(0))])
        embed_lengths, sorted_idx = embed_lengths.sort(0, descending=True)
        embed_index = embed_index[sorted_idx]
        embedded = self.embedding(embed_index)

        packed_input = pack_padded_sequence(embedded, embed_lengths, batch_first=True) 
        packed_output, _ = self.lstm(packed_input)
        hidden, _ = pad_packed_sequence(packed_output, batch_first=True)

        Q = self.w_queue(hidden).to(device)
        K = self.w_key(hidden).to(device)
        V = self.w_value(hidden).to(device)
        dim = Q.size(-1)

        scores = torch.matmul(Q, K.transpose(-2,-1)) / math.sqrt(dim) * 100 # Scalinng up small values to facilitate softmax operation
        attn = torch.matmul(self.dropout(self.relu(F.softmax(scores, dim=-1))), V)
        output, _ = self.lstm2(attn)
        start_output = self.start_proj(output)
        end_output = self.end_proj(output)

        if self.cq:
            for i, idx in enumerate(mask_idx):
                start_output[i][idx:] = -1e9
                end_output[i][idx:] = -1e9

        start_output = F.log_softmax(start_output, dim=1)
        end_output = F.log_softmax(end_output, dim=1)

        return start_output, end_output
````

````python
input_dim = len(train_cq_token_list)
embed_dim = 128
hidden_dim = 128
output_dim = 1
model_sandwich = LSTMSandwich(input_dim, 
                              embed_dim, 
                              hidden_dim,
                              output_dim,
                              bidirectional=True,
                              dropout=0.3,
                              context_question=True).to(device)

NLL = nn.NLLLoss(reduction='sum')
optimizer = torch.optim.Adam(model_sandwich.parameters(), lr=1e-4, weight_decay=0.1)

train_loss_values = []
valid_loss_values = []

for epoch in range(10):
    print('Present Epoch: {}'.format(str(epoch+1) + '/' + str(10)))
    train_loss = 0
    valid_loss = 0

    for a, b, c in tqdm(train_dataloader_cq):
        start_idx = b[:,0].unsqueeze(-1)
        end_idx = b[:,1].unsqueeze(-1)
        
        model_sandwich.train()
        logit1, logit2 = model_sandwich(a.to(device), c.to(device))
        start_loss = NLL(logit1, start_idx.to(device))
        end_loss = NLL(logit2, end_idx.to(device))
        loss = start_loss + end_loss
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    model_sandwich.eval()
    with torch.no_grad():
        for a2, b2, c2 in valid_dataloader_cq: # Because of memory error, I couldn't perform 'full-batch'
            val_start_idx = b2[:,0].unsqueeze(-1)
            val_end_idx = b2[:,1].unsqueeze(-1)

            val_logit1, val_logit2 = model_sandwich(a2.to(device), c2.to(device))
            val_start_loss = NLL(val_logit1, val_start_idx.to(device))
            val_end_loss = NLL(val_logit2, val_end_idx.to(device))
            val_loss = val_start_loss + val_end_loss
            valid_loss += val_loss.item()

        valid_loss_values.append(valid_loss/len(valid_data_processed_cq))

    train_loss_values.append(train_loss/len(train_cq_processed))
    print('Training Average Loss: {}'.format(train_loss_values[-1]))
    print('Valid Average Loss: {}'.format(valid_loss_values[-1]))

plt.plot(train_loss_values, label='train loss')
plt.plot(valid_loss_values, label='valid loss')
plt.xlabel('# of epochs')
plt.ylabel('loss')
plt.legend()
````
